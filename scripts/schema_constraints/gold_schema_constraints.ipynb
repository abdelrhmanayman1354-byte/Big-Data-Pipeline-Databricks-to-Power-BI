## 1. SET COLUMNS TO NOT NULL (Prerequisite for Primary Keys)

```python
%sql
-- For Dimension Tables
ALTER TABLE restaurant_project.gold.dim_products ALTER COLUMN product_skey SET NOT NULL;
ALTER TABLE restaurant_project.gold.dim_branches ALTER COLUMN branch_skey SET NOT NULL;
ALTER TABLE restaurant_project.gold.dim_date     ALTER COLUMN order_date  SET NOT NULL;
ALTER TABLE restaurant_project.gold.dim_payment ALTER COLUMN payment_skey SET NOT NULL;
ALTER TABLE restaurant_project.gold.dim_order_type ALTER COLUMN order_skey SET NOT NULL;
ALTER TABLE restaurant_project.gold.dim_customers ALTER COLUMN customer_id SET NOT NULL;

-- For Fact Table Primary Key
ALTER TABLE restaurant_project.gold.fact_orders  ALTER COLUMN sales_skey SET NOT NULL;
```

## 2. DEFINE PRIMARY KEYS FOR DIMENSION TABLES

```python
%sql
ALTER TABLE restaurant_project.gold.dim_products ADD CONSTRAINT product_pk PRIMARY KEY(product_skey);

ALTER TABLE restaurant_project.gold.dim_branches ADD CONSTRAINT branch_pk PRIMARY KEY(branch_skey);

ALTER TABLE restaurant_project.gold.dim_date ADD CONSTRAINT date_pk PRIMARY KEY(order_date);

ALTER TABLE restaurant_project.gold.dim_payment ADD CONSTRAINT order_pk PRIMARY KEY(payment_skey);

ALTER TABLE restaurant_project.gold.dim_customers ADD CONSTRAINT customer_pk PRIMARY KEY(customer_id);

ALTER TABLE restaurant_project.gold.dim_order_type ADD CONSTRAINT order_type_pk PRIMARY KEY(order_skey);
```

## 3. DEFINE PRIMARY KEY FOR FACT TABLE

```python
%sql
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fact_pk PRIMARY KEY(sales_skey);
```

## 4. DEFINE FOREIGN KEY CONSTRAINTS (Linking Fact to Dimensions)

```python
%sql
-- Link Fact Table to Product Dimension
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fk_product 
FOREIGN KEY(product_skey) REFERENCES restaurant_project.gold.dim_products;

-- Link Fact Table to Branch Dimension
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fk_branch 
FOREIGN KEY(branch_skey) REFERENCES restaurant_project.gold.dim_branches;

-- Link Fact Table to Date Dimension
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fk_date 
FOREIGN KEY(order_date) REFERENCES restaurant_project.gold.dim_date;

-- link Fact Table to Payment Dimension
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fk_payment 
FOREIGN KEY(payment_skey) REFERENCES restaurant_project.gold.dim_payment;

-- link Fact Table to Order Type Dimension
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fk_order_type 
FOREIGN KEY(order_skey) REFERENCES restaurant_project.gold.dim_order_type;

-- link Fact table to customer dimension
ALTER TABLE restaurant_project.gold.fact_orders ADD CONSTRAINT fk_customer 
FOREIGN KEY(customer_id) REFERENCES restaurant_project.gold.dim_customers;
```
